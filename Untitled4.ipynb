```python
import requests
import matplotlib.pyplot as plt

# -------------------------
# 1. Function: Get Calorie Info
# -------------------------
def get_calorie_info(food_name, api_key=None):
    if not api_key:
        # Local fallback calorie database (per 100g)
        local_db = {
            "pizza": 266,
            "hamburger": 250,
            "banana": 89,
            "apple": 52,
            "rice": 130,
            "chicken": 239,
            "pasta": 131
        }
        return local_db.get(food_name.lower(), None)

    # Using USDA FoodData Central API
    url = f"https://api.nal.usda.gov/fdc/v1/foods/search?query={food_name}&api_key={api_key}"
    response = requests.get(url).json()
    try:
        food = response["foods"][0]
        for nutrient in food["foodNutrients"]:
            if nutrient["nutrientName"].lower() == "energy":
                return nutrient["value"]  # kcal per 100g
    except:
        return None

# -------------------------
# 2. Function: Estimate Calories
# -------------------------
def estimate_calories(food_name, portion_size_g=100, api_key=None):
    kcal_per_100g = get_calorie_info(food_name, api_key=api_key)
    if kcal_per_100g:
        total_kcal = (portion_size_g / 100) * kcal_per_100g
        return round(total_kcal, 2), kcal_per_100g
    else:
        return None, None

# -------------------------
# 3. Plotting Function
# -------------------------
def plot_calorie_bar(food_name, portion_size_g, kcal_per_100g, total_kcal):
    labels = ["Per 100g", f"{portion_size_g}g Portion"]
    values = [kcal_per_100g, total_kcal]

    plt.figure(figsize=(5, 4))
    bars = plt.bar(labels, values, color=['#4caf50', '#2196f3'])
    plt.title(f"Calories for {food_name.capitalize()}")
    plt.ylabel("Calories (kcal)")

    # Add values above bars
    for bar, val in zip(bars, values):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
                 f"{val:.1f}", ha='center', va='bottom')

    plt.show()

# -------------------------
# Example Usage
# -------------------------
if __name__ == "__main__":
    API_KEY = None  # Set USDA API key if available
    food_name = "pizza"     # <- Change this to any food name in the database
    portion_size_g = 150    # <- Your portion size in grams

    total_kcal, kcal_per_100g = estimate_calories(food_name, portion_size_g, api_key=API_KEY)

    if total_kcal:
        print(f"{food_name.capitalize()} ({portion_size_g}g): {total_kcal} kcal")
        plot_calorie_bar(food_name, portion_size_g, kcal_per_100g, total_kcal)
    else:
        print("Calorie info not found.")
```
